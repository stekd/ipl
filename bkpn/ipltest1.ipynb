### ipl_deliveries



```python
# Correlation map generated using seaborn

cor_ipl_deliveries = ipl_deliveries.corr()
plt.figure(figsize=(20,10)) # This determines size of the plot
sns.heatmap(cor_ipl_deliveries, annot=True)
plt.show()
```
